## Creating data Frame

```python
df=pd.read_csv("../input/videogamesales/vgsales.csv")
# l=[11011, 11008, 10979, 10022, 10012, 497, 8302, 8292, 8291]
df
```

## What’s the most common publisher?

```python
most_common_publisher = df['Publisher'].mode()
most_common_publisher=most_common_publisher[0]
most_common_publisher
```

## What’s the most common platform?

```python
most_common_platform = df['Platform'].mode()
most_common_platform=most_common_platform[0]
most_common_platform
```

## What’s the most common genre?

```python
most_common_genre = df['Genre'].mode()
most_common_genre=most_common_genre[0]
most_common_genre
```

## What are the top 20 highest grossing games?

```python
top_twenty_highest_grossing_games=df[['Name','Global_Sales']].sort_values(by='Global_Sales',ascending=False)
print(top_twenty_highest_grossing_games.head(20))
```

## For North American video game sales, what’s the median(around median)?

```python
na_median_sales=df['NA_Sales'].median()
# na_median_sales
ten_median_na_seller_names=df[df['NA_Sales']==na_median_sales].sort_values(by='NA_Sales',ascending=False)
# ten_median_na_seller_names
mid_sur=len(ten_median_na_seller_names)//2
ten_median_na_seller_names=list((ten_median_na_seller_names[mid_sur-5:mid_sur+5]).Name)
ten_median_na_seller_names

```

## For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

REF:  https://www.bmj.com/about-bmj/resources-readers/publications/statistics-square-one/2-mean-and-standard-deviation

```python
# top_selling_games=df[['NA_Sales','Global_Sales']].sort_values(by='NA_Sales',ascending=False).head(10)
top_selling_games_na_mean=df['NA_Sales'].mean()
# top_selling_games_na_mean
top_selling_games_na_std=df['NA_Sales'].std()
# top_selling_games_na_std
top_selling_games_na=df.sort_values('NA_Sales', ascending=False)['NA_Sales']
# top_selling_games_na
top_selling_games_all_time=(((top_selling_games_na - top_selling_games_na_mean)//top_selling_games_na_std)).iloc[0]
top_selling_games_all_time
```

## The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?(Multiplicative_difference)!?

```python
wii_sales =df[df['Platform'] == 'Wii']['Global_Sales'].mean()
not_wii_sales=df[df['Platform'] != 'Wii']['Global_Sales'].mean()
multiplicative_difference=wii_sales/not_wii_sales
multiplicative_difference
```

## What are the 5 lowest grossing games?

```python
lowest_sales=df.sort_values('Global_Sales', ascending=False).tail(5).Name
lowest_sales=list(lowest_sales)
lowest_sales
```

## Most CommonYear Which had Most Number of games?

```python
most_common_year = df['Year'].mode()
most_common_year=int(most_common_year[0])
most_common_year
```

## Whats The Newest Game?

```python
latest_game = df[['Name','Year']].max().Name
latest_game
```
